```python
grouped_data = df.groupby(['Sales Channel','Response']).size().unstack(fill_value=0)
```

```python
grouped_data.plot(kind='bar', stacked=False, figsize=(10, 6))
plt.title('Counts of YES and NO for each name in Sales Channel')
plt.xlabel('Name')
plt.ylabel('Counts')
plt.legend(title='Response', loc='upper right', labels=['NO', 'YES'])
plt.xticks(rotation=0)
plt.show()
```
